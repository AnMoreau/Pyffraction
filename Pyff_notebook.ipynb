# Welcome to Pyffraction

Pyffraction is a very simple code to compute diffraction images. For now, only the diffraction figure of holes with a diameter $D$ that are regularly spaced on concentric circles. This figures are simply very nice to watch, but so different from each other that you may want to try yourself ! 

Just download the `Pyffraction.py` (ridiculous, right ?) and run it ! There is an example included, and you can modify the code.

Or you can just import the file once downloaded and play with it. This notebook should be a working example.

We begin with just 12 holes, on a circle with a radius of 1, 10 times the diameter of the holes, then. The code does not check that the holes don't overlap... 

```python
import Pyffraction as pyf
import matplotlib.pyplot as plt

figure = pyf.circular_holes([1],[12])
plt.imshow(figure,cmap = "hot")
plt.axis('off')
plt.show()

```

It's pretty nice, isn't it ? You may want to change the colormap, though, to make it look like it has been produced with a green laser (my favorite). 


```python
import numpy as np
from matplotlib import cm
from matplotlib.colors import ListedColormap
nmax = 256
values = np.ones((nmax, 4))
values[:, 0] = np.linspace(0, 0, nmax)
values[:, 1] = np.linspace(0, 1, nmax)
values[:, 2] = np.linspace(0, 0, nmax)
newcmp = ListedColormap(values)

plt.imshow(figure,cmap = newcmp)
plt.axis('off')
plt.show()

```

Or you may want to use any other colormap, including some that are a bit weird but may end up making you picture quite nice...

```python
plt.imshow(figure,cmap = "cool")
plt.axis('off')
plt.show()
```

Now let's get serious. We want several circular arrays of holes... one of my favorites is 3 rows of circular holes, with 12, 6 and 12 again holes -- on circles that would have 1,2 and 3 as radius. Let's get started ! We need to increase the resolution, though and the size of the figures, to fully enjoy its complexity.

```python
plt.rcParams['figure.dpi'] = 200
figure = pyf.circular_holes([1,2,3],[12,6,12],resolution = 2000)
plt.imshow(figure,cmap = newcmp)
plt.axis('off')
plt.show()
```
